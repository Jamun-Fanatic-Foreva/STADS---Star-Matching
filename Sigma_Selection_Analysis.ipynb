```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
SIGMA = pd.read_csv('F:\IIT Bombay\SatLab\Star Tracker\Programs\CSV Files\SigmaCatalogue.csv')
SIGMA.tail()
```

```python
plt.figure(figsize=(8,6))
plt.hist(SIGMA.mu_n, bins = 40, density = True);
plt.title(r'Histogram of $\mu_n$', fontsize=25);
plt.xlabel(r'$\mu_n \rightarrow$', fontsize=20);
```

```python
global_mu, global_sigma = norm.fit(SIGMA.mu_n)
print('Global mu = {0:2.4f}, Global sigma = {1:2.4f}'.format(global_mu, global_sigma))
k = global_sigma/global_mu
print('Coefficient of Variation = {0:2.6f}'.format(k))
```

```python
x = np.linspace(min(SIGMA.mu_n)-0.05, max(SIGMA.mu_n)+0.05, 1000)
p = norm.pdf(x, global_mu, global_sigma)
```

```python
plt.figure(figsize=(8,6));
plt.hist(SIGMA.mu_n, bins = 50, density = True, alpha=0.6, label = 'Data');
plt.plot(x,p, label = 'Gaussian');
title = r'Fit Results: $\mu$ = {0:2.2f}, $\sigma$ = {1:2.2f}, $k$ = {2:2.3f}'.format(global_mu, global_sigma,k);
plt.title(title, fontsize = 20);
plt.xlabel(r'$\mu_n \rightarrow$', fontsize=15);
plt.legend();
plt.grid();
```

```python
plt.figure(figsize=(8,6));
plt.hist(SIGMA.mu_n, bins = 50, density = True, alpha=0.6, label = 'Data');
plt.plot(x,p, label = 'Gaussian');
title = r'Fit Results: $\mu$ = {0:2.2f}, $\sigma$ = {1:2.2f}, $k$ = {2:2.3f}'.format(global_mu, global_sigma,k);
plt.title(title, fontsize = 20);
plt.xlabel(r'$\mu_n \rightarrow$', fontsize=15);
plt.legend();
plt.axis('equal');
plt.grid();
```
